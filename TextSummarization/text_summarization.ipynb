```python
# prompt template for text summarization
# LLMChain - combine prompt template with llm is known as llm chain (llm | prompt)
from langchain.chains import LLMChain
from langchain import PromptTemplate

generic_prompt = """
Write a summary of the following speech:
Speech:{speech}
Translate the precise summary to: {language}
"""

prompt = PromptTemplate(
    input_variables = ['speech','language'],
    template = generic_prompt 
)
prompt
```

```python
# Stuff document chain text Summarization Technique
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("apjspeech.pdf")
split_docs = loader.load_and_split()
split_docs
#len(split_docs) 7
```

```python
template = """"
Write a concise and short summary of the following speech,
Speech : {text}
"""
prompt = PromptTemplate(
        input_variable=['text'],
        template=template
    )
```

```python
from langchain.chains.summarize import load_summarize_chain
chain = load_summarize_chain(llm = llm, chain_type='stuff', prompt=prompt, verbose=True)
#combined document ,stuff technique will pass my speech from prompt to llm and summarize it
output_summary = chain.invoke(split_docs)
output_summary
```

```python
# map reduce Summarization Technique -> imp for larger files,larger data source
#   -  Multiple prompt template
from langchain.text_splitter import RecursiveCharacterTextSplitter
final_splited_docs = RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(documents=split_docs)
len(final_splited_docs)
```

```python
#from langchain.prompts import ChatPromptTemplate
chunks_template = """"
Please summarize the following speech,
Speech : {text}
Summary:
"""
map_chunks_prompt = PromptTemplate(input_variables=['text'],
                                            template=chunks_template)
```

```python
final_summary_prompt_template = """
Provide the final summary of the entire speech with these important points:
Add a Motivational title. Start the precise summary with an introduction and provide the summary in number points for the speech.
Speech:{text}
"""
final_summary_prompt = PromptTemplate(
                        input_variables=['text'],
                        template=final_summary_prompt_template
                    )
```

```python
# Map Reduce Documents Chain
# map_prompt → the prompt used for each chunk (“map” step).
# combine_prompt → the prompt used to combine the chunk summaries into the final summary.
summary_chain = load_summarize_chain(
                                    llm = llm,
                                    chain_type='map_reduce', 
                                    map_prompt = map_chunks_prompt, 
                                    combine_prompt = final_summary_prompt,
                                    verbose=True
                                    
                                )
# splitted document is more chunked ,map-reduce technique will pass my chunked speech from prompt to llm and summarize it
# then furthuer we will apply another summary prompt over summary chunks to get final summary
map_reduce_summary = summary_chain.invoke(final_splited_docs)
map_reduce_summary
```

```python
# Refine Chain  Summarization Technique
# documents -> chunks -> each chunk passed to prompt then to llm and summarized 
# -> first chunk summary is used with next chunk and then passed to prompt & llm and we get our chunk 2 summary 
# -> iteration for all chunks is done to get final summary

refine_chain = load_summarize_chain(
                                    llm = llm,
                                    chain_type='refine', 
                                    verbose=True
                                    
                                )
# splitted document is more chunked ,map-reduce technique will pass my chunked speech from prompt to llm and summarize it
# then furthuer we will apply another summary prompt over summary chunks to get final summary
refine_summary = refine_chain.invoke(final_splited_docs)
refine_summary
```
